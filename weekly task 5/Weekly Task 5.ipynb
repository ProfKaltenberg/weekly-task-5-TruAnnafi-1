# Question 1

Using the dataset "us-cities-top-1k-multi-year.csv":
1. Create a line graph of total population over time for the states: NY, PA, CT, NJ (10 pts)
2. Be sure to include a title of the graph, labels on your axis, and a legend. (5 pts)
3. Specify colors to be used in your code. (5 pts)

```python
library(ggplot2)
library(dplyr)
library(gridExtra)
```

```python
data <- read.csv("us-cities-top-1k-multi-year.csv") #assigning this csv to data

states_data <- data %>%
    filter(State %in% c("New York", "Pennsylvania", "Connecticut", "New Jersey")) #telling it to only keep the data in those specific states

ggplot(states_data, aes(x = year, y = Population, color = State)) + #starting the graph plus the info from the data columns
  geom_line(linewidth = 1) +
  labs(title = "Total Population Over Time (NY, PA, CT, NJ)", x = "Year", y = "Population", color = "State") + #all the titles and labels
  scale_color_manual(values = c("orange", "blue", "green", "purple")) + #giving the colors I want for each city
  theme_minimal() #gives a cleaner look
```

# Question 2

Using the dataset "us-cities-top-1k-multi-year.csv":
1. Create a Bar Graph of the total population for 5 cities in the state of New York comparing 2014 years and 2018 years (15 pts)
2. Be sure to include a title of the graph, labels on your axis, and a legend. (5 pts)
3. Combine the graph in question 1 and question 2 in one figure.

```python
#filtering the data
ny_data <- data %>%
    filter(State == "New York" & year %in% c(2014, 2018)) %>% #selecting only new york and the years
    group_by(City, year) %>% #grouping the city and year together
    summarise(total_population = sum(Population), .groups = "drop") %>% #gives total population
    slice_head(n = 10)
           
#bar graph data           
ggplot(ny_data, aes(x = City, y = total_population, fill = factor(year))) + #assigning where the data goes and telling to give each year a new color
  geom_bar(stat = "identity", position = "dodge") + #tells to use real values and put the bars side by side
  labs(title = "Population Comparison of NY Cities (2014 vs 2018)", x = "City", y = "Population", fill = "Year") + #chart labels
  scale_fill_manual(values = c("orange", "cyan")) + #saying what colors I want
  theme_minimal() #simple formatting
```
